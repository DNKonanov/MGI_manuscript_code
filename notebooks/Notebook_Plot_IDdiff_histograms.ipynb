#### The code to plot the histogram of ID differences in closely located duplicated reads

```python
from matplotlib.patches import Rectangle

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

with open('../data/duplicate_ID_differences/forward_FOV_data_PE300.txt', 'r') as file:
    
    lines = file.readlines()

data = []
data_by_FOV = {}

for line in lines:
    
    d = line[:-1].split(' ')
    fov = d[0]
    differences = list(map(int, d[1:]))
    
    data += differences
    
    data_by_FOV[fov] = differences
    
    

data = np.array(data)

data = data[data <= 500]


LINECOLOR = 'black'
RECT_COLOR = 'red'
BW_ADJ = 0.3

fig, axes = plt.subplots(2,2, figsize=(14,9))

sns.kdeplot(data, ax = axes[0][0], color=LINECOLOR, linewidth=1, bw_adjust=BW_ADJ)
axes[0][0].set_title('Difference in ID of duplicated reads: PE300, this study')
axes[0][0].set_xlabel('Absolute difference in read ID')
axes[0][0].set_ylim(-0.01, 0.45)


sns.kdeplot(data, ax = axes[0][1], color=LINECOLOR, linewidth=1, bw_adjust=BW_ADJ)
axes[0][1].set_ylim(-0.0001, 0.005)
axes[0][1].set_title('Difference in ID of duplicated reads: PE300, this study\n(Cropped y-scale)')
axes[0][1].set_xlabel('Absolute difference in read ID')



axes[0][1].add_patch(Rectangle((180, 0), 50, 0.0036, edgecolor=RECT_COLOR, fill=False))
axes[0][1].add_patch(Rectangle((50, 0), 50, 0.0018, edgecolor=RECT_COLOR, fill=False))


upline = 0.004
uptext = 0.0042
textdelta = 4

xc = 63
yc = 0.0012
axes[0][1].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[0][1].vlines(ymin=yc, ymax=upline, x=xc, color='gray', linewidth=0.7)


xc = 85
yc = 0.0019
axes[0][1].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[0][1].vlines(ymin=yc, ymax=upline, x=xc, color='gray', linewidth=0.7)


xc = 195
yc = 0.0027
axes[0][1].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[0][1].vlines(ymin=yc, ymax=upline, x=xc, color='gray', linewidth=0.7)


xc = 217
yc = 0.0035
axes[0][1].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[0][1].vlines(ymin=yc, ymax=upline, x=xc, color='gray', linewidth=0.7)




xc = 1
yc = 0.3
axes[0][0].text(xc-textdelta, yc+0.06, f'{xc}', rotation=90)
axes[0][0].vlines(ymin=yc, ymax=yc+0.03, x=xc, color='gray', linewidth=0.7)


uptext=0.13
upline=0.1
downline=0.015
xc = 63
yc = 0.007
axes[0][0].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[0][0].vlines(ymin=downline, ymax=upline, x=xc, color='gray', linewidth=0.7)


xc = 85
yc = 0.007
axes[0][0].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[0][0].vlines(ymin=downline, ymax=upline, x=xc, color='gray', linewidth=0.7)


xc = 195
yc = 0.007
axes[0][0].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[0][0].vlines(ymin=downline, ymax=upline, x=xc, color='gray', linewidth=0.7)


xc = 217
yc = 0.007
axes[0][0].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[0][0].vlines(ymin=downline, ymax=upline, x=xc, color='gray', linewidth=0.7)


with open('../data/duplicate_ID_differences/forward_FOV_data_ext.txt', 'r') as file:
    
    lines = file.readlines()

data = []
data_by_FOV = {}

for line in lines:
    
    d = line[:-1].split(' ')
    fov = d[0]
    differences = list(map(int, d[1:]))
    
    data += differences
    
    data_by_FOV[fov] = differences
    
    

data = np.array(data)

data = data[data <= 500]


LINECOLOR = 'black'
RECT_COLOR = 'red'
BW_ADJ = 0.3

sns.kdeplot(data, ax = axes[1][0], color=LINECOLOR, linewidth=1, bw_adjust=BW_ADJ)
axes[1][0].set_title('Difference in ID of duplicated reads: PE150, CNR0104869')
axes[1][0].set_xlabel('Absolute difference in read ID')
axes[1][0].set_ylim(-0.01, 0.25)


sns.kdeplot(data, ax = axes[1][1], color=LINECOLOR, linewidth=1, bw_adjust=BW_ADJ)
axes[1][1].set_ylim(-0.00005, 0.00025)
axes[1][1].set_title('Difference in ID of duplicated reads: PE150, CNR0104869\n(Cropped y-scale)')
axes[1][1].set_xlabel('Absolute difference in read ID')



#axes[1].add_patch(Rectangle((180, -0.001), 50, 0.029, edgecolor=RECT_COLOR, fill=False))
#axes[1].add_patch(Rectangle((50, -0.001), 50, 0.01, edgecolor=RECT_COLOR, fill=False))



xc = 1
yc = 0.17
axes[1][0].text(xc-textdelta, yc+0.06, f'{xc}', rotation=90)
axes[1][0].vlines(ymin=yc, ymax=yc+0.03, x=xc, color='gray', linewidth=0.7)


uptext=0.13
upline=0.1
downline=0.05
xc = 63
yc = 0.007
axes[1][0].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[1][0].vlines(ymin=downline, ymax=upline, x=xc, color='gray', linewidth=0.7)


xc = 85
yc = 0.007
axes[1][0].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[1][0].vlines(ymin=downline, ymax=upline, x=xc, color='gray', linewidth=0.7)


xc = 195
yc = 0.007
axes[1][0].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[1][0].vlines(ymin=downline, ymax=upline, x=xc, color='gray', linewidth=0.7)


xc = 217
yc = 0.007
axes[1][0].text(xc-textdelta, uptext, f'{xc}', rotation=90)
axes[1][0].vlines(ymin=downline, ymax=upline, x=xc, color='gray', linewidth=0.7)



upline = -0.00001
uptext = -0.000035
textdelta = 4

xc = 63
yc = 0.000025
axes[1][1].text(xc-textdelta, uptext, f'{xc}', rotation=90, color='green')
axes[1][1].vlines(ymin=yc, ymax=upline, x=xc, color='green', linewidth=0.7)


xc = 85
yc = 0.000025
axes[1][1].text(xc-textdelta, uptext, f'{xc}', rotation=90, color='green')
axes[1][1].vlines(ymin=yc, ymax=upline, x=xc, color='green', linewidth=0.7)


xc = 195
yc = 0.000025
axes[1][1].text(xc-textdelta, uptext, f'{xc}', rotation=90, color='green')
axes[1][1].vlines(ymin=yc, ymax=upline, x=xc, color='green', linewidth=0.7)


xc = 217
yc = 0.000025
axes[1][1].text(xc-textdelta, uptext, f'{xc}', rotation=90, color='green')
axes[1][1].vlines(ymin=yc, ymax=upline, x=xc, color='green', linewidth=0.7)



upline = 0.0002
uptext = 0.00022
textdelta = 7

xc = 126
yc = 0.00006
axes[1][1].text(xc-textdelta, uptext, f'{xc}', rotation=90, color='red')
axes[1][1].vlines(ymin=yc, ymax=upline, x=xc, color='red', linewidth=0.7)


xc = 170
yc = 0.00006
axes[1][1].text(xc-textdelta, uptext, f'{xc}', rotation=90, color='red')
axes[1][1].vlines(ymin=yc, ymax=upline, x=xc, color='red', linewidth=0.7)


xc = 390
yc = 0.0001
axes[1][1].text(xc-textdelta, uptext, f'{xc}', rotation=90, color='red')
axes[1][1].vlines(ymin=yc, ymax=upline, x=xc, color='red', linewidth=0.7)


xc = 434
yc = 0.00014
axes[1][1].text(xc-textdelta, uptext, f'{xc}', rotation=90, color='red')
axes[1][1].vlines(ymin=yc, ymax=upline, x=xc, color='red', linewidth=0.7)



plt.tight_layout()

plt.show()
```
